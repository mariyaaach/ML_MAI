### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

````python
import matplotlib.pyplot as plt
````

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df_nagatinski = df[df['sub_area'] == 'Nagatinskij Zaton'].copy()

df_nagatinski.loc[:, 'year'] = df_nagatinski['timestamp'].dt.year

average_price_per_year = df_nagatinski.groupby('year')['price_doc'].mean().reset_index()

plt.figure(figsize=(20, 20))
plt.plot(average_price_per_year['year'], average_price_per_year['price_doc'], marker='o')
plt.title('Average apartment price in Nagatinskij Zaton')
plt.xlabel('Year')
plt.ylabel('Average price')
plt.grid()
plt.xticks(average_price_per_year['year'], rotation=45)
plt.tight_layout()
plt.show()
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

````python
large_apartments = df[df['full_sq'] > 50]

prices = large_apartments['price_doc']

plt.figure(figsize=(10, 6))
plt.hist(prices, bins=50, color='blue', edgecolor='black')
plt.title('Distribution of prices for apartments with an area of ​​more than 50 square meters')
plt.xlabel('Apartment price')
plt.ylabel('Number of apartments')
plt.grid(True)

plt.show()
````

3. Визуализируйте количество проданных квартир в разных районах

````python
sold_apartments_by_area = df['sub_area'].value_counts()

plt.figure(figsize=(48, 48))
plt.bar(sold_apartments_by_area.index, sold_apartments_by_area.values)
plt.title('Number of apartments sold in different areas')
plt.xlabel('Area')
plt.ylabel('Number of apartments')
plt.xticks(rotation=90)
plt.show()
````

4. Постройте box plot отображающий распределение полных площадей квартир в Басманном районе

````python
df_basman = df[df['sub_area'] == 'Basmannoe']

plt.figure(figsize=(20, 20))
plt.boxplot(df_basman['full_sq'], vert=False, patch_artist=True, showmeans=True)
plt.title('Distribution of full area of apartments in Basmannoe district')
plt.xlabel('Full area (sq. meters)')
plt.show()
````

5. Визуализируйте зависимость средней стоимости квартир в зависимости от этажа

````python
avg_price_by_floor = df.groupby('floor')['price_doc'].mean()

plt.figure(figsize=(20, 20))
plt.plot(avg_price_by_floor.index, avg_price_by_floor.values, marker='o')
plt.title('Average Apartment Price vs. Floor')
plt.xlabel('Floor')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
````

6. Покажите как меняется средняя площадь квартир в зависимости от района

````python
avg_area_by_district = df.groupby('sub_area')['full_sq'].mean()

avg_area_by_district = avg_area_by_district.sort_values()

plt.figure(figsize=(48, 48))
plt.bar(avg_area_by_district.index, avg_area_by_district.values)
plt.title('Average apartment area by district')
plt.xlabel('District')
plt.ylabel('Average area (sq. meters)')
plt.xticks(rotation=90)
plt.show()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df_filtered = df[(df['timestamp'].dt.year >= 2011) & (df['timestamp'].dt.year <= 2014) & (df['full_sq'] < 50)]

median_price_by_year = df_filtered.groupby(df_filtered['timestamp'].dt.year)['price_doc'].median()

plt.figure(figsize=(20, 20))
plt.plot(median_price_by_year.index, median_price_by_year.values, marker='o')
plt.title('Median apartment price from 2011 to 2014')
plt.xlabel('Year')
plt.ylabel('Median price')
plt.grid(True)
plt.show()
````
